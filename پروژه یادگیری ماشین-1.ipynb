```python
import numpy as np
import matplotlib.pyplot as plt

# ایجاد دیتاست
X = np.array([[1.2], [2.5], [3.6]])
y = np.array([8.5, 9.2, 10.1])

np.random.seed(0)  # برای تکرارپذیری
w = np.random.rand()  # وزن
b = np.random.rand()  # بایاس

#تابع پیش بینی
def predict(X, w, b):
    return X * w + b

#محاسبه خطا
def quadratic_error(predictions, targets):
    return np.sum((predictions - targets) ** 2)

def absolute_error(predictions, targets):
    return np.sum(np.abs(predictions - targets))

learning_rate = 0.01

for iteration in range(3):
    predictions = predict(X, w, b)

    # محاسبه خطاها
    q_error = quadratic_error(predictions, y)
    a_error = absolute_error(predictions, y)

    # محاسبه گرادیان
    w_gradient = -2 * np.sum((y - predictions) * X) / len(y)
    b_gradient = -2 * np.sum(y - predictions) / len(y)

    # به‌روزرسانی وزن‌ها
    w -= learning_rate * w_gradient
    b -= learning_rate * b_gradient

    print(f"Iteration {iteration + 1}:")
    print(f"Weight (w): {w}, Bias (b): {b}")
    print(f"Quadratic Error: {q_error}, Absolute Error: {a_error}\n")

    # پیش‌بینی نهایی
final_predictions = predict(X, w, b)

# رسم داده‌ها و خط
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, final_predictions, color='red', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
